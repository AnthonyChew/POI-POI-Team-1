```python
import firebase_admin
from firebase_admin import firestore
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="C:/Users/antho/Downloads/poi-poi-f8448-firebase-adminsdk-skqz6-1532cc1a56.json"
import json
from bs4 import BeautifulSoup
from google.cloud.firestore import GeoPoint
import random
import math
```

# Initialize firebase and firestore

```python
app = firebase_admin.initialize_app()
db = firestore.client()
```

# Read json data

```python
all_content = json.load(open('csv/healthier-caterers-geojson.json', encoding='utf-8'))
```

# Data extraction from HTML code

```python
dic = {}
for eatery in all_content['HealthyVendor']:
    soup = BeautifulSoup(eatery['properties']['Description'], 'html.parser')
    tr = soup.findAll('tr')

    dic[eatery['properties']['Name']] = {}
    for t in tr:
        td = t.find('td')
        if(td != None):
            dic[eatery['properties']['Name']][t.find('th').text] = td.text

    latitude = eatery['geometry']['coordinates'][1]
    longitude = eatery['geometry']['coordinates'][0]
    location = GeoPoint(latitude, longitude)

    dic[eatery['properties']['Name']]['coordinates'] = location
    dic[eatery['properties']['Name']]["rating"] =  round(random.uniform(0.0, 5.0) , 1) 
    dic[eatery['properties']['Name']]["likeLocation"] =  []
for data in dic:
    db.collection('healthy_eatery').document(data).set(dic[data])
print("All data has been loaded")
   
```

```python


dic = {}
for data in all_content['features']:
    soup = BeautifulSoup(data['properties']['description'], 'html.parser')

    dic[data['properties']['name']] = {}
    tr = soup.findAll('tr')
    iq = tr[1].findAll('tr') 

    for q in iq :
        h = q.findAll('td')
        dic[data['properties']['name']][h[0].text] = h[1].text

    latitude = data['geometry']['coordinates'][1]
    longitude = data['geometry']['coordinates'][0]
    location = GeoPoint(latitude, longitude)
        
    dic[data['properties']['name']]['coordinates'] = location
    
    
for data in dic:
    db.collection('park').document(data).set(dic[data])
print("All data has been loaded")
```
